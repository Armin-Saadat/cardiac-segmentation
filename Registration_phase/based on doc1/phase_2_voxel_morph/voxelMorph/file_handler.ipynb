```python
import glob
from tqdm import tqdm
import pydicom
import png
import numpy as np
import os
import shutil
```

```python
def divide_into_cine_lge(main_path, cine_path, lge_path):
  i = 0
  cine_counter = 0
  lge_counter = 0
  print()
  for path in tqdm(glob.iglob(main_path + '**/*.png' , recursive=True) , desc = "Reading"):
        file_name = path.split('/')[-1]
        image_type = file_name.split('_')[-1]
        if image_type == 'cine.png':
          if cine_counter > lge_counter:
            print("Error!")
            break
          cine_counter += 1
          des_path = cine_path + str(cine_counter) + '_' + file_name
          i+=1  
        elif image_type == 'affined.png':
          if lge_counter > cine_counter:
            print("Error!")
            break
          lge_counter += 1
          des_path = lge_path + str(lge_counter) + '_' + file_name
          i+=1  
        else: 
          continue
        shutil.copyfile(path, des_path)
        print('number:' , i , "| name:" , path, "copied.")
```

```python
divide_into_cine_lge('./pair_selected/', './Registration_phase/voxelMorph/cine_images/', './Registration_phase/voxelMorph/lge_images/')
```
