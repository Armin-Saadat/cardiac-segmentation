```python
import os
import cv2
import glob
import json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tqdm import tqdm
import SimpleITK as sitk

from scripts.phase5 import *
```

```python
# Reading Dataset
dataset = read_dataset('dataset/ok1,2/1521169/') # cine,lge,mask
```

```python
registered_dataset = Run_simpleITK(dataset , params = [0.01,1e-9,4000])
```

```python
fig ,axes = plt.subplots(len(registered_dataset),5, figsize=(15,25))
for i in range(len(registered_dataset)):
    cine_image , lge_image , mask_image , rigid_lge , rigid_mask, path = registered_dataset[i]
    folder = path.split("/")[3]
    axes[i,0].imshow(lge_image, cmap = 'gray')
    axes[i,0].set_title("main lge" + folder)
    axes[i,1].imshow(cine_image, cmap = 'gray')
    axes[i,1].set_title("main cine" + folder)
    axes[i,2].imshow(rigid_lge, cmap = 'gray')
    axes[i,2].set_title("rigid lge" + folder)
    axes[i,3].imshow(mask_image, cmap = 'gray')
    axes[i,3].set_title("main mask" + folder)
    axes[i,4].imshow(rigid_mask, cmap = 'gray')
    axes[i,4].set_title("rigid mask" + folder)

plt.show()
```
