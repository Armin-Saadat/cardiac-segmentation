# Functions (no need to run)

```python
def eliminate_scar(category, patient_num, sigma=0, save=False, show=True, result_name='scar_eliminated_rigid_lge'):
  
  '''
  get location of scar and substitude intensity from rigid_mask_number.png
  '''

  main_path = './dataset/' + category + '/' + str(patient_num) + '/'
  for folder_num in range(1, 10):
    root = main_path + str(folder_num) + '/results/'
    finished = True
    for path in tqdm(glob.iglob(root + '**/rigid_*.png' , recursive=True) , desc = "Reading"):
      finished = False
      if path.split('_')[-1][0:3] == 'lge':
        rigid_lge_path = path
      elif path.split('_')[-2] == 'mask':
        rigid_mask_path = path
    if finished:
      continue
    rigid_lge = cv2.imread(rigid_lge_path , cv2.IMREAD_GRAYSCALE)
    rigid_mask = cv2.imread(rigid_mask_path, cv2.IMREAD_GRAYSCALE)  
    substitude_intensity = int(rigid_mask_path.split('_')[-1].split('.')[0])

    scar_eliminated_rigid_lge = rigid_lge.copy()
    for i in range(rigid_mask.shape[0]):
      for j in range(rigid_mask.shape[1]):
        if rigid_mask[i][j] > 30:
          if sigma == 0:
            scar_eliminated_rigid_lge[i][j] = substitude_intensity
          else:
            sub = int(np.random.normal(substitude_intensity, sigma, 1)[0])
            if sub > 0 :
              scar_eliminated_rigid_lge[i][j] = sub
            else:
              scar_eliminated_rigid_lge[i][j] = 0

    if show:
      fig , (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 15))
      ax1.imshow(rigid_lge, cmap='gray')
      ax2.imshow(rigid_mask, cmap='gray')
      ax3.imshow(scar_eliminated_rigid_lge, cmap='gray')
      ax1.set_title("Rigid_lge")
      ax2.set_title("Rigid_mask_" + str(substitude_intensity))
      ax3.set_title("Scar_eliminated_rigid_lge")
      
    if save:
      dest = root + result_name + '.png'
      with open(dest, 'wb') as png_file:
        cv2.imwrite(dest, scar_eliminated_rigid_lge)
      print()
      print('saved:', dest)
```

```python
def delete_scar_eliminated_rigid_lge(category, patient_num):
  main_path = './dataset/' + category + '/' + str(patient_num) + '/'
  for folder_num in range(1, 10):
    root = main_path + str(folder_num) + '/results/'
    finished = True
    for path in tqdm(glob.iglob(root + '**/scar_eliminated_rigid_lge.png' , recursive=True) , desc = "Reading"):
      finished = False
      scar_eliminated_rigid_lge_path = path
    if finished:
      continue
    os.remove(scar_eliminated_rigid_lge_path)
    print(scar_eliminated_rigid_lge_path +  ' removed.')
```
